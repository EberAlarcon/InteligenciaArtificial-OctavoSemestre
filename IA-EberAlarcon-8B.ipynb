Ejercicio 30. Representa en una gráfica de barras los valores medios obtenidos en el apartado anterior. Importa los paquetes necesarios para poder representar la gráfica. Añade a la gráfica la información que consideres necesaria para entender mejor los resultados

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de ejemplo
daily_consum = ['BAJO', 'MEDIO', 'ALTO']
valores_promedio = [10, 20, 15]

# Colores personalizados para las barras
colores = ['yellow', 'green', 'red']

# Crear el gráfico de barras
plt.bar(daily_consum, valores_promedio, color=colores)

# Añadir etiquetas a los ejes
plt.xlabel('daily_consum')
plt.ylabel('Valor promedio')

# Añadir título al gráfico
plt.title('Valores promedio de G1, G2 y G3 para los distintos valores posibles de daily_consum')

# Mostrar el gráfico
plt.show()
```
